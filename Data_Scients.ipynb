# Método del codo (Elbow method) para seleccionar el número óptimo de clústeres en K-means

El método del codo es una técnica que se emplea para determinar el número óptimo de clústeres para el algoritmo de K-means. La idea detrás de este método es bastante sencilla. Identificar el número de clústeres para el que se observa un cambio significativo en la tasa de disminución de la varianza intra-cluster (también conocido como suma total de las distancias al cuadrado).

```python
k_values = range(1, 31) # Probar k desde 1 hasta 30
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))

# Graficar los resultados para encontrar el 'codo'
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-')
plt.title('Precisión de K-NN vs. Valor de k (Método del Codo)')
plt.xlabel('Número de Vecinos (k)')
plt.ylabel('Precisión (Accuracy)')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Encontrar el mejor k basado en la mayor precisión
best_k_index = np.argmax(accuracies)
best_k = k_values[best_k_index]
max_accuracy = accuracies[best_k_index]

print(f"\nEl valor de k con la mayor precisión es: {best_k} (Precisión: {max_accuracy:.2f})")
print("Observa la gráfica para identificar el 'codo' que indica un balance entre bias y varianza.")
```
